```python
import pandas as pd
from catboost import CatBoostRanker, Pool
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv(r"E:\Hakaton\participants\participants\data\train_data.csv")
val_data = pd.read_csv(r"E:\Hakaton\participants\participants\data\validation_data.csv")
```

```python
train_data = train_data.sort_values(["user_id", "edition_id"]).reset_index(drop=True)
val_data = val_data.sort_values(["user_id", "edition_id"]).reset_index(drop=True)
```

```python
X_train = train_data.drop(columns=["event_type"])
y_train = train_data["event_type"]
group_train = train_data["user_id"]
```

```python
X_val = val_data.drop(columns=["event_type"])
y_val = val_data["event_type"]
group_val = val_data["user_id"]
```

```python
"""
Обучение для галочки - хуевые негативы, нет cat_features,
нет ре-ранкинга, упирается в переобучение на 86 итерации или около того
А ещё валидация может быть хуйней, но тут вроде с этим все в порядке должно быть

TODO:
Сделать cat_features, нормальные негативы, ре-ранкинг по Diversity
"""


from catboost import Pool

train_pool = Pool(
    data=X_train,
    label=y_train,
    group_id=group_train
)

valid_pool = Pool(
    data=X_val,
    label=y_val,
    group_id=group_val
)

from catboost import CatBoostRanker

model = CatBoostRanker(
    loss_function="YetiRank",
    eval_metric="NDCG:top=20",
    iterations=1000,
    learning_rate=0.05,
    depth=8,
    random_seed=42,
    verbose=100,
    od_type="Iter",
    od_wait=100
)

model.fit(
    train_pool,
    eval_set=valid_pool
)

model.save_model("model.cbm")

"""
Вывод щас:
bestTest = 0.4742697593
bestIteration = 84
"""
```
